```python
def wav2spec(wav_path, downsample=True):
    waveform, sample_rate = torchaudio.load(wav_path)
    if downsample:
        waveform = torchaudio.transforms.Resample(44100, 22050)(waveform)
    specgram = torchaudio.transforms.MelSpectrogram()(waveform)
    return specgram
```

```python
specgram = torchaudio.transforms.MelSpectrogram()(waveform)

print("Shape of spectrogram: {}".format(specgram.size()))

plt.figure()
p = plt.imshow(specgram.log2()[0,:,:].detach().numpy(), cmap='gray')
```

#### repeat and truncate

```python
exp_len = 300 # length on log mel spec
```

```python
def truncate(specgram, exp_len):
    return specgram[:, :, :exp_len]
def repeat(specgram, exp_len):
    num = exp_len//specgram.shape[2]
    tocat = specgram
    for i in range(num - 1):
        tocat = torch.cat((tocat, specgram), dim=2)
    tocat = torch.cat((tocat, specgram[:,:,:(exp_len - specgram.shape[2]*num)]), dim=2)
    return tocat
```

```python
def unify_specgram(specgram, exp_len):
    pin = specgram.shape[2]
    if pin < exp_len:
        return repeat(specgram, exp_len)
    else:
        return truncate(specgram, exp_len)
```

```python
from tensorflow.python.client import device_lib 
print(device_lib.list_local_devices())
```

```python
# import packages
from __future__ import print_function
import numpy as np
import pandas as pd
import tensorflow.keras as keras
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D, GlobalAveragePooling2D
from sklearn.model_selection import train_test_split
import os
```

```python
first_flag = True
for wav_path in train_paths:
    if first_flag:
        to_add = unify_specgram(wav2spec(wav_path), exp_len)
        first_flag = False
    else:
        to_add = torch.cat((to_add, unify_specgram(wav2spec(wav_path), exp_len)), dim=0)
print(to_add.shape)
```

```python
first_flag = True
for wav_path in devel_paths:
    if first_flag:
        dev = unify_specgram(wav2spec(wav_path), exp_len)
        first_flag = False
    else:
        dev = torch.cat((dev, unify_specgram(wav2spec(wav_path), exp_len)), dim=0)
print(dev.shape)
```

```python
X_shape = to_add.shape
X_train = to_add.numpy().reshape(X_shape[0], X_shape[1], X_shape[2], 1)
X_devel = dev.numpy().reshape(dev.shape[0], dev.shape[1], dev.shape[2], 1)
# X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.05, random_state=42)
label_file    = '../lab/labels.csv'   # path of labels
df_labels = pd.read_csv(label_file)
y_train = df_labels['label'][df_labels['file_name'].str.startswith('train')].values
y_devel = df_labels['label'][df_labels['file_name'].str.startswith('devel')].values
print('X_train shape:', X_train.shape)
print(X_train.shape[0], 'train samples')
print(X_devel.shape[0], 'test samples')
```

```python
np.save('X_train.npy', X_train)
np.save('X_devel.npy', X_devel)
```

```python
# Encode labels (necessary to obtain decision function output)
from sklearn.preprocessing import LabelEncoder
classes   = ['noise','orca']
encoder = LabelEncoder()
encoder.fit( classes )
y_train = encoder.transform(y_train)
y_devel = encoder.transform(y_devel)
```

```python
y_train
```

```python
# transform label into one-hot vector
y_train = keras.utils.to_categorical(y_train, 2)
y_devel = keras.utils.to_categorical(y_devel, 2)
```

```python
y_train
```

```python
# configuration
config = {'batch_size':32,
          'num_classes':2,
          'epochs':50,
          'model_n':'SGD',
          'data_augmentation':False,
          'model_dir':os.path.join(os.getcwd(), 'saved_models/'),
          'npy_dir':os.path.join(os.getcwd(), 'saved_npys/'),
          'csv_dir':os.path.join(os.getcwd(), 'saved_csvs/')
         }

batch_size = config['batch_size']
num_classes = config['num_classes']
epochs = config['epochs']
model_n = config['model_n']
data_augmentation = config['data_augmentation']
model_dir = config['model_dir']
if not os.path.isdir(model_dir):
    os.makedirs(model_dir)
npy_dir = config['npy_dir']
if not os.path.isdir(npy_dir):
    os.makedirs(npy_dir)
csv_dir = config['csv_dir']
if not os.path.isdir(csv_dir):
    os.makedirs(csv_dir)
model_name = 'CNN_%d_epochs_%d_batch_%s.h5'%(epochs, batch_size, model_n)
npy_name = 'CNN_%d_epochs_%d_batch_%s.npy'%(epochs, batch_size, model_n)
```

```python
model = Sequential()
model.add(Conv2D(64, kernel_size=(5,5),
                 activation='relu',
                 input_shape=(to_add.shape[1],to_add.shape[2],1)))
model.add(MaxPooling2D(pool_size=(2, 2),strides=2))
model.add(Dropout(0.25))

model.add(Conv2D(128, (5,5),activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2),strides=2))
model.add(Dropout(0.25))

model.add(GlobalAveragePooling2D())

model.add(Dense(1024, activation='relu'))
model.add(Dense(2, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(),
              metrics=['accuracy'])
print(model.summary())
```

```python
# set early stop condition
from keras.callbacks import EarlyStopping
early_stopping_monitor = EarlyStopping(monitor='val_loss', patience=100)

# set checkpoint save best model
check_pointer = keras.callbacks.callbacks.ModelCheckpoint(model_dir+model_name, monitor='val_accuracy', 
                                          verbose=1, save_best_only=True, save_weights_only=True, mode='max', period=1)
# initialize optimizer
optimizers = {'rmsprop':keras.optimizers.rmsprop(lr=0.0001, decay=1e-5),
              'Adadelta':keras.optimizers.Adadelta(learning_rate=0.01, rho=0.95, epsilon=1e-6),
              'SGD':keras.optimizers.SGD(learning_rate=0.001, momentum=0.9, nesterov=True)
             }
opt = optimizers[model_n]
# Use optimizer to train model
model.compile(loss='categorical_crossentropy',
              optimizer=opt,
              metrics=['accuracy'])

# process the training data
X_train = X_train.astype('float32')
X_devel = X_devel.astype('float32')
```
